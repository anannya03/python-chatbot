```python
import json

f= open('train-v2.0.json')
data= json.load(f)

ans_dict={}
a= data["data"]
for b in a:
    para=b['paragraphs']
    for quas in para:
        quas_1= quas['qas']
        for q in quas_1:
            questions= q['question']
            answers= q['answers']
            if(len(answers)==0):
                break
            ans= answers[0]
            main_answer= ans['text']
            questions= questions.lower()
            main_answer= main_answer.lower()
            ans_dict[questions]= main_answer
            
```
